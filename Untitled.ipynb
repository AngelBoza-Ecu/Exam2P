## Import libraries

```python
import numpy as np
import streamlit as st
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import plotly_express as px
from scipy import stats
from streamlit_option_menu import option_menu
```

## Read File

```python
well_file = "Volve production data.xlsx"
data = pd.read_excel(well_file)
print(data)
```

```python
curve = data['BORE_OIL_VOL']
print(curve)
```

```python
# Creating Dataframe
Volo = data['BORE_OIL_VOL']
#Volg = data['BORE_GAS_VOL']
time = data['DATEPRD']
fig, ax = plt.subplots(figsize=(20, 10))
x = time
y = Volo
#Build the curve
ax.plot(x, y, c='g')
ax.set_xlabel('t(años)', fontsize=14)
ax.set_ylabel('Volo', fontsize=14)
ax.set_title('Volo vs t(años)', fontsize=18)

ax.grid()
plt.show()
```
